````python
# -------------------------------
# Task Manager (Weeks 4–9 style)
# Prints doctest results AND a demo run like your screenshot
# -------------------------------

tasks = []  # in-memory list

def add_task(task):
    """
    Add a non-blank task and print confirmation.

    >>> tasks[:] = []
    >>> add_task("Buy milk")
    Added: Buy milk
    >>> add_task("   ")
    Task cannot be blank
    >>> tasks
    ['Buy milk']
    """
    task = task.strip()
    if task == "":
        print("Task cannot be blank")
        return
    tasks.append(task)
    print(f"Added: {task}")

def remove_task(index):
    """
    Remove by 0-based index with bounds check.

    >>> tasks[:] = ["A", "B", "C"]
    >>> remove_task(1)
    Removed: B
    >>> tasks
    ['A', 'C']
    >>> remove_task(5)
    Invalid index!
    >>> remove_task(-1)
    Invalid index!
    """
    if 0 <= index < len(tasks):
        removed = tasks[index]
        del tasks[index]
        print(f"Removed: {removed}")
    else:
        print("Invalid index!")

def list_tasks():
    """
    Print tasks using 1-based numbering or a message if empty.

    >>> tasks[:] = []
    >>> list_tasks()
    No tasks available.
    >>> tasks[:] = ["Buy milk", "Pay bills"]
    >>> list_tasks()
    1: Buy milk
    2: Pay bills
    """
    if not tasks:
        print("No tasks available.")
    else:
        for i, t in enumerate(tasks, start=1):
            print(f"{i}: {t}")

def main():
    # Demo output like your photo
    add_task("Buy milk")
    add_task("Pay bills")
    add_task("Walk dog")

    print("\nCurrent Tasks:")
    list_tasks()

    print("\nRemoving task at index 1:")
    remove_task(1)

    print("\nTasks after removal:")
    list_tasks()

if __name__ == "__main__":
    # Run doctests (Week 9) then the demo
    import doctest
    doctest.testmod(verbose=True)
    main()
````
